```python
#Calculadora Financiera Simple
#Funciones:
#-Interés simple: Calcula el interés basado en el capital, la tasa de interés y el tiempo.
#-Interés compuesto: Calcula el valor futuro con interés compuesto.
#-Valor presente: Calcula el valor presente de un valor futuro dado.
#-Valor futuro: Calcula el valor futuro de un capital con interés simple.
#-Pagos de préstamo: Calcula los pagos periódicos de un préstamo usando la fórmula de anualidades.

def menu():
    print("\nCalculadora Financiera")
    print("1. Interés Simple")
    print("2. Interés Compuesto")
    print("3. Valor Presente")
    print("4. Valor Futuro")
    print("5. Pago de Préstamo")
    print("6. Salir")

def interes_simple(principal, tasa, tiempo):
    return principal * (1 + tasa * tiempo)

def interes_compuesto(principal, tasa, tiempo, compuestos_por_año):
    return principal * (1 + tasa / compuestos_por_año) ** (compuestos_por_año * tiempo)

def valor_presente(valor_futuro, tasa, tiempo):
    return valor_futuro / (1 + tasa) ** tiempo

def valor_futuro(principal, tasa, tiempo):
    return principal * (1 + tasa) ** tiempo

def pago_prestamo(principal, tasa, num_pagos):
    return (principal * tasa) / (1 - (1 + tasa) ** -num_pagos)

def main():
    while True:
        menu()
        opcion = input("Seleccione una opción: ")

        if opcion == '1':
            # Interés Simple
            principal = float(input("Capital inicial: "))
            tasa = float(input("Tasa de interés anual (como decimal, ej. 0.05 para 5%): "))
            tiempo = float(input("Tiempo (en años): "))
            total = interes_simple(principal, tasa, tiempo)
            print(f"El monto total con interés simple es: {total:.2f}")

        elif opcion == '2':
            # Interés Compuesto
            principal = float(input("Capital inicial: "))
            tasa = float(input("Tasa de interés anual (como decimal): "))
            tiempo = float(input("Tiempo (en años): "))
            compuestos_por_año = int(input("Número de veces que se capitaliza por año: "))
            total = interes_compuesto(principal, tasa, tiempo, compuestos_por_año)
            print(f"El monto total con interés compuesto es: {total:.2f}")

        elif opcion == '3':
            # Valor Presente
            valor_futuro = float(input("Valor futuro: "))
            tasa = float(input("Tasa de descuento anual (como decimal): "))
            tiempo = float(input("Tiempo (en años): "))
            vp = valor_presente(valor_futuro, tasa, tiempo)
            print(f"El valor presente es: {vp:.2f}")

        elif opcion == '4':
            # Valor Futuro
            principal = float(input("Capital inicial: "))
            tasa = float(input("Tasa de interés anual (como decimal): "))
            tiempo = float(input("Tiempo (en años): "))
            vf = valor_futuro(principal, tasa, tiempo)
            print(f"El valor futuro es: {vf:.2f}")

        elif opcion == '5':
            # Pago de Préstamo
            principal = float(input("Monto del préstamo: "))
            tasa_anual = float(input("Tasa de interés anual (como decimal): "))
            num_anios = int(input("Número de años para pagar: "))
            num_pagos = num_anios * 12  # Asumiendo pagos mensuales
            tasa_mensual = tasa_anual / 12
            pago = pago_prestamo(principal, tasa_mensual, num_pagos)
            print(f"El pago mensual es: {pago:.2f}")

        elif opcion == '6':
            print("Saliendo de la calculadora financiera. ¡Gracias por usarla!")
            break

        else:
            print("Opción no válida. Por favor, intente de nuevo.")

if __name__ == "__main__":
    main()
```
