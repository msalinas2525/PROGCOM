JUEGO DE MATEMÁTICA BÁSICA CON FUNCIONES

```python
import random
import time

# Funciones para realizar las operaciones básicas
def suma(x, y):
    return x + y

def resta(x, y):
    return x - y

def multiplicacion(x, y):
    return x * y

def division(x, y):
    if y == 0:
        return "Indefinido"
    else:
        return round(x / y, 2)  # Redondeamos la división a 2 decimales

def pregunta_suma():
    num1, num2 = random.randint(1, 100), random.randint(1, 100)
    respuesta_correcta = suma(num1, num2)
    return f"{num1} + {num2}", respuesta_correcta

def pregunta_resta():
    num1, num2 = random.randint(1, 100), random.randint(1, 100)
    respuesta_correcta = resta(num1, num2)
    return f"{num1} - {num2}", respuesta_correcta

def pregunta_multiplicacion():
    num1, num2 = random.randint(1, 12), random.randint(1, 12)
    respuesta_correcta = multiplicacion(num1, num2)
    return f"{num1} * {num2}", respuesta_correcta

def pregunta_division():
    num1, num2 = random.randint(1, 100), random.randint(1, 12)
    respuesta_correcta = division(num1, num2)
    return f"{num1} / {num2}", respuesta_correcta

# Función para generar preguntas de operación aleatoria (mezcladas)
def pregunta_mezclada():
    tipo_operacion = random.choice(['suma', 'resta', 'multiplicacion', 'division'])
    if tipo_operacion == 'suma':
        return pregunta_suma()
    elif tipo_operacion == 'resta':
        return pregunta_resta()
    elif tipo_operacion == 'multiplicacion':
        return pregunta_multiplicacion()
    else:
        return pregunta_division()

def menu():
    print("\n--- Menú de operaciones ---")
    print("1. Suma")
    print("2. Resta")
    print("3. Multiplicación")
    print("4. División")
    print("5. Mezcladas")
    print("6. Salir")

# Función principal del juego
def juego_operaciones():
    puntaje = 0  # Contador de respuestas correctas

    while puntaje < 10:
        menu()
        opcion = input("Elige una opción (1/2/3/4/5/6): ")

        if opcion == '6':
            print("Has salido del juego.")
            break
        if opcion == '1':
            pregunta, respuesta_correcta = pregunta_suma()
        elif opcion == '2':
            pregunta, respuesta_correcta = pregunta_resta()
        elif opcion == '3':
            pregunta, respuesta_correcta = pregunta_multiplicacion()
        elif opcion == '4':
            pregunta, respuesta_correcta = pregunta_division()
        elif opcion == '5':
            pregunta, respuesta_correcta = pregunta_mezclada()
        else:
            print("Opción inválida. Por favor elige una opción correcta.")
            continue

        # Mostrar la pregunta y medir el tiempo de respuesta
        print(f"\nResponde: {pregunta}")
        start_time = time.time()  # Inicia el cronómetro

        try:
            respuesta_usuario = input("Respuesta (o escribe 'salir' para terminar): ")
            if respuesta_usuario.lower() == 'salir':
                print("Has salido del juego.")
                break

            # Verificar el tiempo de respuesta
            tiempo_transcurrido = time.time() - start_time
            if tiempo_transcurrido > 10:
                print(f"¡Se acabó el tiempo! Tardaste {int(tiempo_transcurrido)} segundos.")
                continue

            # Comparar la respuesta del usuario con la respuesta correcta
            if float(respuesta_usuario) == respuesta_correcta:
                puntaje += 1
                print(f"¡Correcto! Tu puntaje actual es: {puntaje}/10")
            else:
                print(f"Incorrecto. La respuesta correcta era: {respuesta_correcta}")
        except ValueError:
            print("Respuesta inválida. Por favor ingresa un número o 'salir'.")

    if puntaje == 10:
        print("¡Felicidades! Has ganado el juego.")

# Ejecutar el juego
juego_operaciones()
```
